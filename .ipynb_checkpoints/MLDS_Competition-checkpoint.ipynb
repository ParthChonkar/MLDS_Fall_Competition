<h2> Models </h2>

Logistic Regression Model

```python
class simple_logreg(nn.Module):
    
    def __init__(self,inputs):
        super(simple_logreg, self).__init__()
        self.linear = nn.Linear(inputs,1)
        
    def forward(self,x):
        return torch.sigmoid(self.linear(x))
```

Neural Network with 1 hidden layer


```python
class NeuralNet(nn.Module):
    
    def __init__(self, inputSize, h1Size, outputSize):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(inputSize,h1Size)
        self.linear2 = nn.Linear(h1Size,outputSize)

        
    def forward(self,x):
        activationLayer1 = torch.sigmoid(self.linear1(x))
        outputLayer = self.linear2(activationLayer1)
        return torch.sigmoid(outputLayer)
    
net = NeuralNet(2,2,2)
t1 = torch.tensor([1.,2.])
pred = net(t1)
print(pred)
```

Neural Network with 2 hidden layers

```python
class NeuralNet2(nn.Module):
    
    def __init__(self, inputSize, h1Size, h2Size, outputSize):
        super(NeuralNet2, self).__init__()
        self.linear1 = nn.Linear(inputSize,h1Size)
        self.linear2 = nn.Linear(h1Size,h2Size)
        self.linear3 = nn.Linear(h2Size,outputSize)
        
    def forward(self,x):
        activationLayer1 = torch.sigmoid(self.linear(x))
        activationLayer2 = torch.sigmoid(self.linear(activationLayer1))
        outputLayer = self.linear3(activationLayer2)
        return outputLayer
```

Pytorch optimizers and cost functions + model instantiation

```python
### THINGS THAT WILL BE USED IN TRAINING ##

trainingX = scaled_TrainX.float()
trainingY = scaled_TrainY.float()
logreg_Model = simple_logreg(25)
logreg_Optimizer = torch.optim.SGD(logreg_Model.parameters(), lr=1, momentum = 0.5)
costFunc = nn.BCELoss()
```

<h2> Training  </h2>
Hyper Parameter Log:

* Model1 1: logreg (lr=1 m=0.5 parameters [0:24])


```python
def LGR_stochastic_gradientDescent(epochs):
    for epoch in range(epochs):
        totalCost = 0
        for i in range(129):
            logreg_Optimizer.zero_grad()
            pred = logreg_Model(trainingX[i]) #CHANGE BACK TO TRAINING
            loss = costFunc(pred,trainingY[i])
            totalCost+=loss
            loss.backward()
            logreg_Optimizer.step()
        if(epoch%500==0):
            print("Cost: ", totalCost)
```

```python
# created NN compatible y tensors 
nnY = []
for x in range(scaled_TrainY.shape[0]):
    nnY.append([1 - scaled_TrainY[x],0 + scaled_TrainY[x]])
    
scaled_Train_NNY = torch.tensor(nnY).float()
trainingX = scaled_TrainX.float()
nn_Model = NeuralNet(25, 5, 2 )
nn_Optimizer = torch.optim.SGD(nn_Model.parameters(), lr=1, momentum = 0.5)
nn_costFunc = nn.BCELoss()
costs = []
```

```python
#for training on val data
nnY=[]
for x in range(scaled_ValidationY.shape[0]):
    nnY.append([1 - scaled_ValidationY[x],0 + scaled_ValidationY[x]])
scaledNNY = torch.tensor(nnY).float()
```

```python
costs = []
def nn_SGD(epochs):
    for epoch in range(epochs):
        totalCost = 0
        for i in range(129):
            nn_Optimizer.zero_grad()
            pred = nn_Model(trainingX[i])
            loss = nn_costFunc(pred,scaled_Train_NNY[i])
            totalCost+=loss
            loss.backward()
            nn_Optimizer.step()
        if(epoch%10==0):
            costs.append(float(totalCost))
        if(epoch%500==0):
            print("Cost: ", totalCost)
```

Load Saved Models

```python
def predict(model,X):
    prediction = model(X)
    
    if(len(model.state_dict())==4):
        predictions = []
        for x in range(X.size()[0]):
            predictions.append(prediction[x].argmax())       
        prediction = torch.tensor(predictions)

    elif(len(model.state_dict())==2):   
        for x in range(X.size()[0]):
            if(prediction[x]>=0.5):
                prediction[x] = 1
            else:
                prediction[x] = 0
    return prediction
```

```python
def testValidationData(model, val_dataX, val_dataY):
    true_pos = 0
    true_neg = 0
    false_pos = 0
    false_neg = 0
    total = 0
    
    validation_Prediction = predict(model,val_dataX.float())
           
    
    for x in range(val_dataY.size()[0]):
        if ((validation_Prediction[x].float()== 1) and (val_dataY[x].float() ==1)):
            true_pos = true_pos + 1
        elif ((validation_Prediction[x].float()== 0) and (val_dataY[x].float() ==0)):
            true_neg = true_neg + 1
        elif ((validation_Prediction[x].float()== 1) and (val_dataY[x].float() ==0)):
            false_pos = false_pos + 1
        else:
            false_neg = false_neg + 1 
        
        total = total + 1
        
    accuracy = (true_pos + true_neg)/total
    precision = true_pos/(true_pos + false_pos)
    recall = true_pos/(true_pos + false_neg)
    print("Validation Data Accuracy: ", accuracy)
    print("Validation Data Precision: ", precision)
    print("Validation Data Recall: ", recall)
    print("F1 Score: ", (precision * recall * 2)/(precision+recall) )
    print(true_pos)
```

Run Validation Tests on loaded models and Check Precision, Recall, Accuracy and F scores 

```python
#Model1 - Logistic Regression
print("Logistic Regression Model Validation Statistics: ")
testValidationData(savedModel1, scaled_ValidationX, scaled_ValidationY)
```
